#Weather Advisor

```python
Temp = float(input("Enter the current Temp in Celsius: "))
```

```python
if Temp >= 30:
    print("It's a hot day. Stay hydrated!")
elif Temp >= 20:
    print("It's a warm day. Enjoy the weather!") # This covers 20.00 to 29.99
elif Temp >= 10:
    print("It's a cool day. Wear a jacket!")  # This covers 10.00 to 19.99
else:
    print("It's a cold day. Stay warm!")  # This covers anything below 10.00
```

#Numbers Divisible

```python
numbers = list(range(1, 21))
```

```python
for num in numbers:
    if num % 3 == 0:
        print("Numbers divisible by 3 is:", num)
```

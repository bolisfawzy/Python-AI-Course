# Total Budget

```python
Budget= 60
```

```python
print("My Budget  is", Budget)
print("Type of Budget:", type(Budget))
```

#Cost Calculation

```python
Cost= Item1+Item2+Item3
```

```python
print("Total Cost= ", Cost)
print("Type of Cost:", type(Cost))
```

```python
print(f"Total cost of items: ${Cost:.2f}")
print(f"Budget: ${Budget:.2f}")

if Cost <= Budget:
    remaining = Budget - Cost
    print(f"You are within Budget.")
    print(f"Money left: ${remaining:.2f}")
else:
    needed = Cost - Budget
    print(f"You have exceeded the Budget.")
    print(f"Additional money needed: ${needed:.2f}")
```
